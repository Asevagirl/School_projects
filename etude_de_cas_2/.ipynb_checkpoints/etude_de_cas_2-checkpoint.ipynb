```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFolder = '/home/johan/Documents/Data/EpEn data/'
```

```python
nb_ad = 28
nb_mci = 22
nb_sci = 22
idx_freq = ['ALPHA', 'BETA', 'DELTA', 'THETA']
idx_seuil = [100, 70, 50, 30, 20, 10]
```

```python
#Création de 3 listes de dictionnaires (un pour chaque patient de chaque type).
#Les dictionnaires créés ici contienent la matrice d'asynchronisation pour toutes les fréquences
x_ad = []
for i in range(1, nb_ad+1):
    x = {}
    for freq in idx_freq:
        x[freq + '_100'] = loadmat(dataFolder + freq + '/' 
                                    + 'AD' + '/' 
                                    + str(i) + '.mat')[freq]
    x_ad.append(x)

x_mci = []
for i in range(1, nb_mci+1):
    x = {}
    for freq in idx_freq:
        x[freq + '_100'] = loadmat(dataFolder + freq + '/' 
                                    + 'MCI' + '/' 
                                    + str(i) + '.mat')[freq]
    x_mci.append(x)

x_sci = []
for i in range(1, nb_sci+1):
    x = {}
    for freq in idx_freq:
        x[freq + '_100'] = loadmat(dataFolder + freq + '/' 
                                    + 'SCI' + '/' 
                                    + str(i) + '.mat')[freq]
    x_sci.append(x)
```

```python
#On étend les dictionaires aux 6 différents niveau de seuil
def get_percentile(x, seuil):
    y = x
    percentile_limit = np.percentile(x, 100 - seuil)
    y[y<percentile_limit] = 0
    return y


for i in range(nb_ad):
    for freq in idx_freq:
        for seuil in idx_seuil:
            x_ad[i][freq+'_'+str(seuil)] = get_percentile(x_ad[i][str(freq)+'_100'], seuil)
            
for i in range(nb_mci):
    for freq in idx_freq:
        for seuil in idx_seuil:
            x_mci[i][freq+'_'+str(seuil)] = get_percentile(x_mci[i][str(freq)+'_100'], seuil)
            
for i in range(nb_sci):
    for freq in idx_freq:
        for seuil in idx_seuil:
            x_sci[i][freq+'_'+str(seuil)] = get_percentile(x_sci[i][str(freq)+'_100'], seuil)
```

```python
#On se ramène de 30 mesures par éléctrodes à une mesure par électrode
def nb_connexions_electrode(array):
    connexions = []
    for i in range(array.shape[0]):
        connexion = 0
        for j in range(len(connexions)):
            if array[i, j]>0:
                connexion += 1
        connexions.append(connexion)
    return connexions

#Il faudrait peut etre essayer en enlevant les connexions qui ont disparues, et calculer une vraie CDF
def variance_cdf(array):
    variance_list = []
    for i in range(array.shape[0]):
        cdf = np.cumsum(array[i, :])
        variance_list.append(np.var(cdf))
    return variance_list
```
